```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
text_data = df['source_name'] + ' ' + df['source_name_hi'] + ' ' + df['source_code']
```

```python
# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(text_data)

# Calculate cosine similarity
cos_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
print(cos_sim)
```

```python
threshold = 0.9  # You can adjust this threshold based on your data
# Check for delays based on cosine similarity
for i in range(len(cos_sim)):
    for j in range(i+1, len(cos_sim)):
        if cos_sim[i][j] < threshold:
            print(f"Train from {df['source_name'][i]} to {df['source_name'][j]} is delayed.")
            break
        else:
            print(f"No delay detected for train from {df['source_name'][i]} to {df['source_name'][j]}.")
```

```python

num_instances = 5

# Calculate the number of comparisons made during cosine similarity calculation
num_comparisons = num_instances * (num_instances - 1) // 2

# Define the probability of selecting 1 (delayed) and 0 (not delayed)
probability_1 = 0.9  # Adjust this probability as needed

# Generate random ground truth delay values (1s and 0s) with more 1's
ground_truth_delays = [random.choices([0, 1], weights=[1-probability_1, probability_1])[0] for _ in range(num_comparisons)]

print("Randomly generated ground truth delay values:")
print(ground_truth_delays)
```

```python
# Initialize a list to store the detected delays
detected_delays = []

# Check for delays based on cosine similarity
for i in range(len(cos_sim)):
    for j in range(i+1, len(cos_sim)):
        if cos_sim[i][j] < threshold:
            detected_delays.append(1)  # Delayed
        else:
            detected_delays.append(0)  # Not delayed

# Calculate accuracy by comparing detected delays with ground truth delays
correct_predictions = sum(detected_delay == ground_truth_delay for detected_delay, ground_truth_delay in zip(detected_delays, ground_truth_delays))
accuracy = correct_predictions / len(ground_truth_delays)

print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Predictions from your delay detection algorithm
# Replace this with the actual predictions
predicted_delays = [random.choice([0, 1]) for _ in range(num_comparisons)]

# Calculate precision, recall, and F1-score
precision = precision_score(ground_truth_delays, predicted_delays)
recall = recall_score(ground_truth_delays, predicted_delays)
f1 = f1_score(ground_truth_delays, predicted_delays)

print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
